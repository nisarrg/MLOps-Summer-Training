```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
```

```python
male = cv2.imread('male.jpg')
female = cv2.imread('female.jpg')
```

```python
male = cv2.resize(male, (512, 512))
```

```python
female = cv2.resize(female, (512, 512))
```

```python
cv2.imshow('Photo', male)
cv2.waitKey()
cv2.destroyAllWindows()


cv2.imshow('Photo', female)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(female)
```

```python
plt.imshow(male)
```

```python
cmale = male[50:300, 150:300]
```

```python
female[50:300, 100:250] = cmale
```

```python
cv2.imshow('Swapped Image', female)
cv2.waitKey()
cv2.destroyAllWindows()
```
