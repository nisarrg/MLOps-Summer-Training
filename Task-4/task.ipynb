```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt
```

```python
w, h = 512, 512
arr = np.zeros((h, w, 3), dtype=np.uint8)
cv2.line(arr, (126,156), (126, 306), thickness=3, color=(0, 0, 255))
cv2.line(arr, (126, 306), (226, 306), thickness = 3, color=(0, 0, 255))
cv2.line(arr, (236, 156), (286, 306), thickness = 3, color=(0, 0, 255))
cv2.line(arr, (286, 306), (326, 186), thickness = 3, color=(0, 0, 255))
cv2.line(arr, (326, 186), (386, 306), thickness = 3, color=(0, 0, 255))
cv2.line(arr, (386, 306), (436, 156), thickness = 3, color=(0, 0, 255))
plt.imshow(arr)
```

```python
cv2.imwrite('LW.png', arr)
```
